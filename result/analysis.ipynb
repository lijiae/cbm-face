```python
target_attributions= ['Male','Young', 'Middle_Aged', 'Senior','Asian','White','Black','Rosy_Cheeks','Shiny_Skin','Bald','Wavy_Hair','Receding_Hairline','Bangs','Sideburns',
                'Black_Hair','Blond_Hair','Brown_Hair','Gray_Hair','No_Beard','Mustache',
                '5_o_Clock_Shadow','Goatee','Oval_Face','Square_Face','Round_Face','Double_Chin',
                'High_Cheekbones','Chubby','Obstructed_Forehead', 'Fully_Visible_Forehead','Brown_Eyes',
                'Bags_Under_Eyes','Bushy_Eyebrows','Arched_Eyebrows','Mouth_Closed','Smiling',
                'Big_Lips','Big_Nose','Pointy_Nose','Heavy_Makeup','Wearing_Hat','Wearing_Earrings',
                'Wearing_Necktie','Wearing_Lipstick','No_Eyewear','Eyeglasses','Attractive']
```

```python
def plt_attr(a1,a2,a3):
    plt.figure(figsize=(20,30))
    font={
        'weight':'normal',
        'size':20,
    }
    plt.xlabel('acc',font)
    plt.plot(a1.tolist(),target_attributions,linewidth=3,marker='o',markersize=10)
    plt.plot(a2.tolist(),target_attributions,linewidth=3,marker='o',markersize=10)
    plt.plot(a3.tolist(),target_attributions,linewidth=3,marker='x',markersize=10)
    # plt.barh(range(len(target_attributions)),a3.tolist())
    plt.yticks(size=15)

    plt.legend(['modified-0','modified-1','prediction'])
    plt.savefig("acc.png")
    plt.show()
```

```python
plt_attr(result_m_0_gt,result_m_1_gt,acc)
```

```python
def draw_score(mini):
    plt.figure(figsize=(20,30))
    font={
        'weight':'normal',
        'size':20,
    }
    plt.xlabel('attribution',font)
    plt.barh(target_attributions,mini.tolist())
    plt.yticks(size=15)
    # plt.xticks(target_attributions)

    plt.legend(['m-1 - m-0'])
    plt.savefig("mini.png")
    plt.show()
```

```python
draw_score(result_m_1_gt-result_m_0_gt)
```
